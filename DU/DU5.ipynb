## Sparse matice (sousednosti)
Vytvořte sparse matici o rozměrech $n^2\times n^2$. Tato matice reprezentuje pravidelnou čtvercovou síť o velikosti $n \times n$. Každá buňka této čtvercové sítě odpovídá jednomu řádku/sloupci matice, přičemž pořadí buněk čtvercové sítě berte po řádcích (viz obrázek níže).

Hodnoty v matici budou řízeny jedním parametrem $\varepsilon$; pro jednoduchost si místo něj představte hodnotu $0.1$.
Nenulové hodnoty v matici jsou zadány sousedností buněk následovně:
- hodnoty na diagonále (tedy sousednost prvku se sebou samým) jsou $1 + k \cdot \varepsilon$ (kde $k$ určuje počet sousedních buněk: 4 uvnitř, 3 na hranách, 2 v rozích),
- mimodiagonální prvky mají hodnotu $-\varepsilon$ právě tehdy, pokud index řádku a index sloupce mimodiagonálního prvku odpovídá sousedním buňkám ve čtvercové síti (viz obrázek).

**Ukázka, jak přechází mřížka do matice pro $n=4$ a $\varepsilon = 0.1$, je na následujícím obrázku:**


Vznikne matice s následujícími vlastnostmi:
- velikost $n^2 \times n^2$,
- $\varepsilon$ je parametr, který můžete nastavit, například $\varepsilon = 0.1$,
- matice je symetrická, tj. $A = A^T$,
- každý řádek/sloupec má maximálně 5 nenulových prvků,
    - 5, pokud se jedná o buňku uvnitř sítě,
    - 4, pokud se jedná o buňku na okraji sítě,
    - 3, pokud se jedná o buňku v rohu sítě,
- výsledná matice je pětidiagonální (hlavní diagonála, dvě vedlejší těsně pod a nad hlavní diagonálou, dvě vedlejší diagonály ve vzdálenosti $\pm n$ od hlavní diagonály).

## Úkol 1
**Sestavte matici pomocí formátu `LIL` a vyplňte hodnoty pomocí cyklu přes všechny buňky čtvercové sítě.**
- vypište a vykreslete ji jako obrázek pro $n=3$ a $\varepsilon = 0.1$,
- vykreslete ji jako obrázek pro $n=100$ a $\varepsilon = 0.2$.

````python
import scipy.sparse as sparse
import numpy as np

def A_assembly(n=3, epsilon=1e-1):
    A = sparse.lil_matrix((n*n, n*n))
    for i in range(n):
        for j in range(n):
            index = i * n + j

            sousedi = 0

            if i > 0:
                sousedi += 1
            if i < n - 1:
                sousedi += 1
            if j > 0:
                sousedi += 1
            if j < n - 1:
                sousedi += 1

            A[index, index] = 1 + epsilon * sousedi

            if i > 0:
                A[index, (i-1)*n + j] = -epsilon
            if i < n - 1:
                A[index, (i+1)*n + j] = -epsilon
            if j < 0:
                A[index, i*n + (j-1)] = -epsilon
            if j < n-1:
                A[index, i*n + (j+1)] = -epsilon

    return A

A = A_assembly(3, 1e-1)
print(A.todense())
````

````python
import matplotlib.pyplot as plt

plt.imshow(A.todense())
plt.colorbar()
plt.show()
````

## Bonusový úkol

**Sestavte matici $A$ z úkolu 1 bez použití cyklů.**
- porovnejte výsledky z úkolu 1 a 2 pro různé hodnoty $n$ a $\varepsilon$.

Nejprve sestavte vektory indexů řádků, sloupců a hodnot pro všechny nenulové prvky matice. Poté sestavte sparse matici ve formátu CSR najednou z těchto tří polí.

Hint, jak na to:
- Sestavte tři pole `row`, `col` a `data`, která obsahují řádky, sloupce a hodnoty nenulových prvků matice.
    - separátně sestavte příspěvky, které vzniknou konektivitou na vnitřních horizontálních hranách a na vnitřních vertikálních hranách,
        - při procházení všech vnitřních horizontálních hran si rozmyslete, jaké jsou vždy indexy nad a pod hranou (to jsou souřadnice řádků a sloupců nenulových prvků),
        - totéž pro vnitřní vertikální hrany; zde se může hodit `np.arange(n * n).reshape((n, n)).T.flatten()`, což vytvoří vektor s přeházenými indexy (tedy indexy buněk uspořádané po sloupcích),
    - zvlášť sestavte příspěvky pro mimodiagonální prvky a diagonální prvky,
        - u diagonálních prvků nejprve vyrobte vektor hodnot $1 + 4\cdot \varepsilon$ a pak pro každou hranu odečtěte $\varepsilon$.
- Tato pole vytvořte vektorizovaně, bez cyklů.

````python
# create matrix A using vectorized operations, creating indices for rows and columns and data
# use sparse.csr_matrix((data, (i, j)))
import numpy as np
from scipy.sparse import csr_matrix

n = 3
epsilon = 0.1

def A_efficiently(n, epsilon):
    ## --- Mimodiagonální prvky
    # Vnitřní horizontální hrany
    horizontal_up_idx = np.arange(n*(n-1)).reshape(n-1, n).flatten()
    horizontal_low_idx = horizontal_up_idx + n

    row_horizontal = np.concatenate([horizontal_up_idx, horizontal_low_idx])
    col_horizontal = np.concatenate([horizontal_low_idx, horizontal_up_idx])
    data_horizontal = -epsilon * np.ones(2 * (n * n - n)) # data jsou vždy stejná


    # Vnitřní vertikální hrany
    index_matrix_vertical = np.arange(n * n).reshape((n, n)).T.flatten() # indexy buněk přeházené pro vertikální hranu
    vertical_left_idx = index_matrix_vertical.reshape(n, n)[:, :-1].flatten()
    vertical_right_idx = index_matrix_vertical.reshape(n, n)[:, 1:].flatten()

    row_vertical = np.concatenate([vertical_left_idx, vertical_right_idx])
    col_vertical = np.concatenate([vertical_right_idx, vertical_left_idx])
    data_vertical = -epsilon * np.ones(2 * (n * n - n)) # data jsou vždy stejná

    ## --- Diagonála (všechny prvky)
    diagonal_idx = np.arange(n * n)
    row_idx_diag = diagonal_idx
    col_idx_diag = diagonal_idx
    data_diag = (1 + 4 * epsilon) * np.ones(n * n)

    # odečteme epsilon pro každou z hran
    # horní hrana
    top = np.arange(n)
    # dolní hrana
    bot = np.arange(n*(n-1), n*n)
    # levá hrana
    left = np.arange(0, n*n, n)
    # pravá hrana
    right = np.arange(n-1, n*n, n)

    #odečteme epsilon pro každou z hran
    data_diag = data_diag.copy()

    for idx in top:
        data_diag[idx] -= epsilon
    for idx in bot:
        data_diag[idx] -= epsilon
    for idx in left:
        data_diag[idx] -= epsilon
    for idx in right:
        data_diag[idx] -= epsilon

    ## --- spojení všech vektorů a sestavení matice
    # spojíme všechny vnitřní hrany a diagonálu
    row = np.hstack((row_idx_diag, row_horizontal, row_vertical))
    col = np.hstack((col_idx_diag, col_horizontal, col_vertical))
    data = np.hstack((data_diag, data_horizontal, data_vertical))

    # vyrobíme matici
    A_efficiently = csr_matrix((data, (row, col)), shape=(n*n, n*n))

    return A_efficiently

print(A_efficiently(n, epsilon).todense())
````
