1. Vytvořte funkci, která vrací faktoriál přirozeného čísla.

````python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
````

4. Vytvořte funkci, která najde prvních `n` prvočísel.

````python
def first_primes(n):
    primes = []
    num = 2
    while len(primes) < n:
        is_prime = True
        for p in primes:
            if num % p == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
        num += 1
    return primes

print(first_primes(10))
````

# 2. Ladění

**Euklidův algoritmus pro výpočet největšího společného dělitele (NSD) dvou čísel**

```text
Mějme dvě přirozená čísla uložená v proměnných u a w.
Dokud w není 0, opakuj:
    r = u % w
    u = w
    w = r
Po skončení cyklu je v u uložen NSD původních čísel.
```

## 2.1 Úkoly

- Nastavte breakpoint na řádek 3, spusťte buňku v debug režimu a sledujte změny hodnot proměnných `u` a `w` během celého algoritmu.
- Vytvořte podmíněný breakpoint na řádku 5, který se zastaví až ve chvíli, kdy bude `w == 0`.
- Nastavte logovací breakpoint na řádku 5, který vypíše hodnoty proměnných `u` a `w`.


````python
def gcd(a, b):
    # Euklidův algoritmus pro výpočet největšího společného dělitele (NSD) dvou čísel
    u = a
    w = b
    while w != 0:
        r = u % w
        u = w
        w = r
    return u


a = 2 * 2 * 3 * 5 * 7 * 11 * 13
b = 2 * 3 * 3 * 7 * 17

print(a)
print(b)

gcd(a, b)
````
